```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0,30,1000)
Y = np.zeros(1000)

#resize picture size
fig = plt.figure(figsize=(10,10))

#draw all boundry
for num in range(0,31):
    plt.plot(Y+num,X,'b')
    plt.plot(X,Y+num,'b')
    

fx =np.zeros(5)
fy =np.linspace(0,1,5)

#symmetric
for factor in [-1,1]:
    #column that need to be fill
    for col in [2,5,7,10,11,18,25]:
        #fill the color
        for num in range(0,50):
            plt.plot(factor*(fx+num*0.02+0)+15,fy+col,'k')


```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0,32,1000)
Y = np.zeros(1000)

#resize picture size
fig = plt.figure(figsize=(10,10))

#draw all boundry
for num in range(0,33):
    plt.plot(Y+num,X,'b')
    plt.plot(X,Y+num,'b')
    

fx =np.zeros(5)
fy =np.linspace(0,1,5)

Blackfill_list = [[23,24],[22,25],[22,25],[4,5,6,17,18,22,26,27],[4,7,18,21,27],[2,3,7,18,20,24,27],[1,6,18,19,23,25,26]\
                  ,[1,7,15,16,18,19,22],[2,3,4,8,13,14,18,21],[5,8,12,18,22],[5,6,9,11,18,23],[4,5,8,9,10,13,14,18,21,24]\
                  ,[3,7,9,12,13,14,15,18,22,24],[3,6,7,8,12,13,14,15,18,25],[2,6,8,13,14,18,25],[2,5,7,10,11,18,25]]

#symmetric
for factor in [-1,1]:
    #column that need to be fill
    for col in range(0,16):
        for row in Blackfill_list[col]:
        #fill the color
            for num in range(0,8):
                plt.plot(factor*(fx+num*0.125+15-col)+16,fy+row,'k')
```
