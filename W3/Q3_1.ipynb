```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import signal
%matplotlib inline

#rgbgrayr g b convolution
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray



#3*3 Gassian filter
gaussian_kernel = [[1,2,1],[2,4,2],[1,2,1]]
gaussian_kernel = gaussian_kernel/ np.array([16,16,16])
#5*5 Gassian filter
gaussian_kernel2 = [[1,4,7,4,1],[4,16,26,16,4],[7,26,41,26,7],[4,16,26,16,4],[1,4,7,4,1]]
gaussian_kernel2 = gaussian_kernel2/ np.array([273,273,273,273,273])


#Main function
image = mpimg.imread('./image.jpg')
gray = rgb2gray(image) #RGB to gray
#
plt.imshow(gray, cmap=plt.get_cmap('gray'))#grayscale
plt.axis('off')
plt.show()#show

con3_3 = signal.convolve2d(gray, gaussian_kernel) #
plt.axis('off')
plt.imshow(con3_3, cmap=plt.get_cmap('gray'))
plt.show()#show

con5_5 = signal.convolve2d(gray, gaussian_kernel2) #
plt.axis('off')
plt.imshow(con5_5, cmap=plt.get_cmap('gray'))
plt.show()#show
```
