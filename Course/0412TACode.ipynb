```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
print(boston)
```

```python
x_data = boston.data
y_data = boston.target
```

```python
x_data[0]
```

```python
y_data[0]
```

```python
boston.feature_names
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_data,y_data,test_size = 0.3,random_state = 2)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X =iris.data
```

```python
X[0]
```

```python
Y = iris.target
```

```python
Y[0]
```

```python
Y
```

```python
x_data = X[:,1:3]
```

```python
x_data[0]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_data,Y,test_size = 0.2,random_state = 87)
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_train) # x的兩個feature是x y軸 顏色是y的分類
```

```python
clf = SVC()
clf.fit(x_train,y_train)
```

```python
y_predict =clf.predict(x_test)
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_predict)
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_predict)#看看命中率多少
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_predict)#預測是0 實際也是0有13個，預測是1 實際也是1有8個，預測是2 實際也是2有7個
```
