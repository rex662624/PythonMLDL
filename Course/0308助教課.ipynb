- %matplotlib inline 是用來即時顯示圖片

```python
import matplotlib.pyplot as plt
%matplotlib inline
x = [1,2,3,4]
y1 = [2,3,1,4]
y2 = [5,2,4,9]
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(5,5),facecolor="b")
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,y1,label = "y1")
plt.plot(x,y2,label = "y2")
plt.legend(loc ="best")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.title("title")
```

```python
import numpy as np

plt.figure(figsize=(5,5))
plt.plot(x,y1,label = "y1")
plt.plot(x,y2,label = "y2")
plt.legend(loc ="best")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.title("title")

plt.xlim(-2,20)
plt.ylim(-20,10)
plt.xticks(np.arange(-2,20,5));#-2到50每隔5格

plt.yticks(np.arange(-20,10,3));#-2到50每隔5格
```
