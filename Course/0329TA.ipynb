### groupby

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo_state = ufo.groupby('State') #會把State的部分合併起來 
```

```python
ufo_state #只是標籤，不是能印出來的datatype
```

```python
ufo_state.count()#city 116 , colors 25 ,it shows there is many NAN in the dataframe.And count won't count it
```

```python
ufo_state.sum()
```

```python
ufo_state 
```

```python
ufo_state = ufo.groupby('State').count()
ufo_state
```

```python
ufo_state.sort_values(by='Time',ascending=False,inplace =True)
```

```python
ufo_state[:10].Time.plot(kind= 'bar')
```

```python
ufo.groupby(['State','City']).count()
```

```python
ufo.groupby(['State','City']).count().reset_index()
```

### datetime

```python
ufo['datetime']=pd.to_datetime(ufo.Time)#把Time轉成共同格式的datetime
```

```python
ufo.head()
```

```python
import datetime as dt
```

```python
ufo['month']= ufo.datetime.dt.month
```

```python
ufo.head()
```

```python
ufo['weekday']=ufo.datetime.dt.weekday_name
```

```python
ufo.head()#因為Datetime已經認得時間了，所以他知道是星期幾
```

```python
ufo_weekdays =ufo.groupby("weekday").count()
```

```python
ufo_weekdays.sort_values(by = 'Time',
                        ascending = False,
                        inplace = True)
```

```python
ufo_weekdays.Time.plot(kind='bar')
```
